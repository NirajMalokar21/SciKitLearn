# Decision tree 

### The decision tree starts with distinct classes compiled at the root node, and then it produces decision boundaries or branches to partition the items into their respective classes at the leaf nodes.

### It makes no assumption about any data, knnown as CART (Classification and Regression Tree)

### Performance does not get limited by non linear data

--- 

## Disadvantages

### Deccision trees can easily overfit when tree is too complex or when there is noise or irrelevant data in the dataset

### It can be biased towards features that appear earlier in the tree or have more splits

### Small variations in data can lead to different tree structures

### Not well suited for complex problems with continuous data, works best for categorical or discreet data

---


## Definitions

### Entropy is the measure of randomness or unpredictability in the dataset (usually between 0 and 1. If higher than one dat has high amount of disorder). Entropy is calculated at each stage to decide if it needs to continue
### Information gain represents how much entropy was removed during splitting at a node. Entropy(T) is entropy at node before split, Entropy(Tv) is entropies after split (Where T and Tv are number of instances)

### Gini Impurity Calculates the purity of split. When the leaf notes are the same its gini impurity is 0.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
data = load_wine()
print(data.keys())
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df.info()
```

```python
target = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, target, test_size=0.20, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
csf = DecisionTreeClassifier(max_depth=3, random_state=42)

csf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = csf.predict(X_test)

print("Accuracy score: ", accuracy_score(y_test, y_pred))
```

## How does it make the decision as to what the split will be?

### This is what it thinks about during the training
### At each step, the model will choose the split that maximises information gain
### We measure information gain using entropy (formula is sum( -Pi log Pi))
### Information gain = E(parent) - sum ( Wi * E(CHILDi))
### Model tries every possible split, and chooses the one with the most IG
### Model stops trying when gini impurity is 0

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15,10))
plot_tree(csf, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.show()
```
